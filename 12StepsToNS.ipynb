```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
#matplotlib inline

#variables declaration
nx = 35
ny = 35
nt = 17
dx = 2 / (nx - 1)
dy = 2 / (ny - 1)

sigma = 0.25
mu = 0.05
dt = sigma * dx * dy / mu

def diffuse(nt):
    #initial condition
    u = np.ones((ny, nx))
    un = np.ones((ny, nx))
    u[int(0.5/dy):int(1/dy+1), int(0.5/dx):int(1/dx+1)] = 2

    x = np.linspace(0, 2, nx)
    y = np.linspace(0, 2, ny)
    X, Y = np.meshgrid(x, y)

    #fig1 = plt.figure(figsize = (11,7), dpi = 100)
    #ax1 = fig1.gca(projection = '3d')
    #ax1.plot_surface(X, Y, u, rstride = 2, cstride = 2, cmap = cm.viridis)
    #ax1.set_xlabel('X')
    #ax1.set_ylabel('Y')
    #ax1.set_zlabel('U')
    for n in range(nt+1):
        un = u.copy()
        u[1:-1, 1:-1] = un[1:-1, 1:-1] + \
                    mu * dt / dx**2 * (un[1:-1, 2:] - 2 * un[1:-1, 1:-1] + un[1:-1, :-2]) + \
                    mu * dt / dy**2 * (un[2:, 1:-1] - 2 * un[1:-1, 1:-1] + un[:-2, 1:-1])
        u[0, :] = 1
        u[:, 0] = 1
        u[-1, :] = 1
        u[:, -1] = 1
    
    fig2 = plt.figure(figsize = (11,7), dpi = 100)
    ax2 = fig2.gca(projection = '3d')
    ax2.plot_surface(X, Y, u, rstride = 2, cstride = 2, cmap = cm.viridis)
    ax2.set_xlabel('X')
    ax2.set_ylabel('Y')
    ax2.set_zlabel('U')
    
```
